```
#Soal 1
#Input jari-jari dan tinggi 
r=float(input("Input Jari-jari= "))
t=float(input("Input Tinggi= "))
#phi menggunakan math.pi
phi=mt.pi
#Perhitungan
Lp=2*phi*r**2+2*phi*r*t
#Output
print("Output:",Lp)
```

```
#Soal2
#Inputan angka
bil = int(input("Input bilangan: "))
#if parent
if bil > 1:
  for i in range(2, bil):
    if (bil % i) == 0:
      print("Output:", bil," Bukan bilangan prima")
      break
  else:
    print("Output:", bil," Adalah bilangan prima")
else:
  print("Output:", bil," Bukan bilangan prima")
```

```
#Soal3
kata=str(input("Input Kata:"))
if(kata==kata[::-1]):
  print("Output: "+kata+" Adalah palindrome")
else:
  print("Output: "+kata+" Bukan palindrome")
```

```
#Soal4
a = int(input("Input: "))
z = a - 1 
for i in range(0, a):
    for j in range(0, z):
        print(' ', end='')
    z -= 1
    for j in range(0, i + 1):
        print('* ', end='')
    print('')
```
